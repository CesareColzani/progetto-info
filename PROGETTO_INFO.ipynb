```python
#5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.
```

```python
#uso tabella gia creata dove prendevo i valori per year piu recente
new=times.iloc[times.groupby('university_name')['year'].idxmax()]
new.head(3)
```

```python
#separo il valore male-female e li metto in 2 colonne differenti 
new["rappfemmine"]=new["female_male_ratio"].str.split(" : ",expand=True)[0]
new["rappmaschi"]=new["female_male_ratio"].str.split(": ",expand=True)[1]
new.head(3)
```

```python
#trasformo valori femmine e maschi in valori numerici(pd.to_numeric()):
new['rappfemmine']=pd.to_numeric(new['rappfemmine'])
new['rappmaschi']=pd.to_numeric(new['rappmaschi'])
new.head(3)
```

```python
# moltiplico numero tot studenti per il rapporto maschi-femmine 
new['femmine']=(new['rappfemmine']*new['num_students'])/100
new['maschi']=(new['rappmaschi']*new['num_students'])/100
new.head(3)
```

```python
#con l' operazione di prima ho ottenuto il numero di studenti raggruppati per ogni universita, ma mi serve raggruppati per country 
new.groupby('country').sum()['maschi']
```

```python
new.groupby('country').sum()['femmine']
```

```python
#6. Find the universities where the ratio between female and male is below the average ratio (computed over all universities)
```

```python
#uso tabella con gia group by fatto, anche se prendo ultimio anno, il rapporto negli anni per stessa università non varia
# in questo modo avro una sola università anche se come gia detto il rapporto è lo stesso 
#calcolo rapporto tra maschi e femmine per ogni scuola(piu veloce piuttosto che farlo sul numero di studenti ) 

new['school_ratio']=new['rappfemmine']/new['rappmaschi']
new
```

```python
#sommo il rapporto femmine 
somma_rappf=new['rappfemmine'].sum()
somma_rappf
```

```python
#sommo rapporto maschi
somma_rappm=new['rappmaschi'].sum()
somma_rappm
```

```python
#divido per ottenere il rapporto totale tra femmine e maschi 
average_ratio_tot=somma_rappf/somma_rappm
average_ratio_tot
```

```python
uni=set(new[new['school_ratio']<average_ratio_tot]['university_name'])
uni
```

```python
#7. For each country, compute the fraction of the students in the country that are in one of the universities computed in the previous point (that is, the denominator of the ratio is the total number of students over all universities in the country).
```

```python
x=new[new['school_ratio']<average_ratio_tot].groupby('country')['num_students'].sum()
x
```
